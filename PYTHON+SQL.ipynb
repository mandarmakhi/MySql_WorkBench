```python
ps.sqldf("""

SELECT 
    Embarked,
    Sex,
    AVG(Age) AS avg_age,
    SUM(Age) AS tot_age,
    MIN(Age) AS min_age,
    MAX(Age) AS max_age,
    COUNT(PassengerId) AS tot_passengers,
    COUNT(DISTINCT Ticket) AS dist_tickets
    
FROM df AS f

WHERE age >= 18
    AND age <=60
    
GROUP BY Embarked,
    Sex

HAVING avg_age > 30

ORDER BY Embarked DESC,
    Sex DESC

""")
```
